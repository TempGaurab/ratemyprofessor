```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
with open("informatics.TXT", "r", encoding="utf-8") as file:
    content = file.read()
```

```python
content
```

```python
soup = BeautifulSoup(content, parser = "html.parser")
soup
```

```python
soup.prettify()
```

```python
target_div = soup.find("div", class_="TeacherFilterSet__StyledTeacherFilterSet-c31t21-0 gfzvFL")
target_div
```

```python
next_div = target_div.find_next_sibling("div")
next_div
```

```python
a_tags = soup.find_all("a", class_="TeacherCard__StyledTeacherCard-syjs0d-0 dLJIlx")
a_tags
```

```python
data = []
for a_tag in a_tags:
    href = a_tag.get("href")
    link = "https://www.ratemyprofessors.com" + href #works
    div = a_tag.find("div", class_ = "CardName__StyledCardName-sc-1gyrgim-0 cJdVEK")
    name = div.text #works
    rating_div = a_tag.find("div", class_=lambda x: x and 'CardNumRating__CardNumRatingNumber' in x) 
    rating_text = rating_div.text if rating_div else '0'    #works
    rating_count_div = a_tag.find("div", class_=lambda x: x and 'CardNumRating__CardNumRatingCount' in x) 
    rating_count = rating_count_div.text if rating_div else '0'
    Course_div = a_tag.find("div", class_=lambda x: x and 'CardSchool__Department' in x) 
    course_text = Course_div.text if Course_div else 'Not Provided' #works
    feedback_divs = a_tag.find_all("div", class_=lambda x: x and 'CardFeedback__CardFeedbackNumber' in x)
    would_take_again = feedback_divs[0].text if len(feedback_divs) > 0 else '0'
    difficulty = feedback_divs[1].text if len(feedback_divs) > 1 else '0'
    University = "Northern Kentucky University" #works
    data.append({
        'Name': name,
        'Rating Value': rating_text,
        'Rating Count': rating_count,
        'Course': course_text,
        'Would Take Again': would_take_again,
        'Difficulty': difficulty,
        'University': University,
         'Link': link
    })
```

```python
df = pd.DataFrame(data)
df
```

```python
df["Rating Count"] = df["Rating Count"].str . replace("ratings", "")
df
```

```python
#rating value
#rating count
#would take again
#difficulty
df["Rating Count"] = df["Rating Count"].astype(int)
filtered_df = df[df['Rating Count'] != 0]
filtered_df
```

```python
df['Rating Value'] = df['Rating Value'].astype(float)
df['Difficulty'] = df['Difficulty'].astype(float)
df['Would Take Again'] = df['Would Take Again'].replace('N/A', '0%').str.rstrip('%').astype(float)
```

```python
#for a professor to be good:
# high rating value (most important)
#low diffuculty
# high would take again but(if it 0% i dont want the formula to emss up)
#rating count ( but make sure 0 wont affect it)
```

```python
df['Normalized Rating Value'] = (df['Rating Value'] - df['Rating Value'].min()) / (df['Rating Value'].max() - df['Rating Value'].min())

# Normalize Difficulty (invert so lower values are better)
df['Normalized Difficulty'] = 1 - (df['Difficulty'] - df['Difficulty'].min()) / (df['Difficulty'].max() - df['Difficulty'].min())

# Normalize Rating Count
k = 1  # Small constant to avoid division by zero
df['Normalized Rating Count'] = df['Rating Count'] / (df['Rating Count'] + k)

# Handle Would Take Again
df['Would Take Again'] = df['Would Take Again'] / 100
```

```python
df
```

```python
# Calculate Score
w1, w2, w3, w4 = 0.3, 0.3, 0.1, 0.3
df['Score'] = (w1 * df['Normalized Rating Value'] +
               w2 * df['Normalized Difficulty'] +
               w3 * df['Would Take Again'] +
               w4 * df['Normalized Rating Count'])
```

```python
df
```

```python
df_sorted = df.sort_values(by='Score', ascending=False)
df_sorted.drop( columns = ["Normalized Rating Value", "Normalized Difficulty", "Normalized Rating Count"],inplace = True)
df_sorted
```

```python
pd.set_option('display.max_colwidth', None)  # Show full column width
df_sorted.reset_index(inplace= True)
df_sorted
```
