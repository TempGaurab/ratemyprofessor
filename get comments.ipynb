```python
import glob
import pandas as pd
from bs4 import BeautifulSoup
```

```python
all_names = []
all_comments = []
all_departments = []
```

```python
df = pd.DataFrame(columns=['Name', 'Comment', 'Department'])
df
```

```python
def scrape(raw_html):
    soup = BeautifulSoup(raw_html, 'html.parser')
    names_elements = soup.find_all(class_='NameTitle__Name-dowf0z-0 cfjPUG')  # Replace with the actual class for names
    
    comments_elements = soup.find_all(class_='Comments__StyledComments-dzzyvm-0')
    departments_elements = soup.find_all(class_='TeacherDepartment__StyledDepartmentLink-fl79e8-0')  # Replace with the actual class for departments
    
    names = [name.get_text() for name in names_elements]
    comments = [comment.get_text() for comment in comments_elements]
    departments = [dept.get_text() for dept in departments_elements]
    return names, comments, departments
```

```python
# Path to the root folder where your .txt files are located
path_to_files = './'  # Adjust this if your files are in a different directory

# Pattern to match all page_*.txt files
file_pattern = f'{path_to_files}page_*.txt'

# List all files matching the pattern
file_list = sorted(glob.glob(file_pattern), key=lambda x: int(x.split('_')[-1].split('.')[0]))
```

```python
# Iterate through each file in the list
for file_path in file_list:
    try:
        # Open and read the file
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            
            # Process the content with the scrape function
            names, comments, departments = scrape(content)
    
            # Append results to the main lists
            all_names.extend(names)
            all_comments.append(comments)
            all_departments.extend(departments)
            
    except Exception as e:
        print(f"An error occurred while reading {file_path}: {e}")
```
